```python
!pip install biopython -q
from Bio.Seq import Seq
import random

# --- TASK 2: Utility Class ---
# Task 2: Create a utility class, which contains both methods as static methods.
# This class acts as a container for our methods.
# "Static" means we can call them without creating a 'new BioUtils()' object.
class BioUtils:
    @staticmethod
    def transcribe_dna_to_rna(dna):
        result = dna.transcribe()
        return result

    @staticmethod
    def translate_rna_to_protein(rna):
        result = rna.translate()
        return result

# --- TASK 3 & 4: SequenceStorage & Singleton Pattern ---
# Task 4: Change the class SequenceStorage in a way that only one object could exist.
# Design Pattern Used: Singleton Pattern. It ensures that only one instance of the class can be created.
class SequenceStorage:
    _instance = None

    def __new__(cls):
        # If no instance exists yet, create it. Otherwise, return the existing one.
        if cls._instance is None:
            cls._instance = super(SequenceStorage, cls).__new__(cls)
            cls._instance.data = {}
        return cls._instance

    def save(self, name, seq):
        self.data[name] = seq

    def read(self, name):
        return self.data[name]

# --- TASK 6 & 7: New Requirements & Factory Pattern ---
# Task 6: Extend the code to also work with protein sequences.
# These classes allow the program to handle different sequence types.
class DNASequence:
    def __init__(self, sequence):
        self.seq = Seq(sequence)

class ProteinSequence:
    def __init__(self, sequence):
        self.seq = Seq(sequence)

# Task 7: Create a class SequenceFactory, which contains a method to create a sequence.
# Based on a given flag, the sequence will be a DNA or protein sequence.
class SequenceFactory:
    @staticmethod
    def create_sequence(type_flag, sequence_string):
        if type_flag == "DNA":
            return DNASequence(sequence_string)
        elif type_flag == "PROTEIN":
            return ProteinSequence(sequence_string)

class DNASequenceGenerator:
    alphabet = ['A','C','G','T']
    def create_sequence(self, n):
        result = ''
        for i in range(n):
            idx = random.randint(0,3)
            result = result + DNASequenceGenerator.alphabet[idx]
        return result

# --- MAIN EXECUTION ---
if __name__ == '__main__':

    # Task 1: Translate the given DNA sequence into a protein sequence.
    sequence = 'GTGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG'
    dna_obj = Seq(sequence)

    # Using the Utility Class methods (from Task 2)
    rna = BioUtils.transcribe_dna_to_rna(dna_obj)
    protein = BioUtils.translate_rna_to_protein(rna)

    # Task 3: Store the sequence into an object of the class SequenceStorage.
    storage = SequenceStorage()
    storage.save("OriginalProtein", protein)

    # Task 5: Take the coding sequence of the gene CD28. What is the corresponding protein sequence?
    # This is the human CD28 CDS (Coding Sequence)
    cd28_str = "ATGATCCTCAGGCTGGGCTTCTGCCTGGCTCTGGTTCTGGGGGTTCTGGGAGCCATTTGGGCTGGGCCTGGGGAGAGTCTCCTGGTCTACCCACTGAAG"
    cd28_dna = Seq(cd28_str)
    cd28_protein = BioUtils.translate_rna_to_protein(BioUtils.transcribe_dna_to_rna(cd28_dna))
    storage.save("CD28_Protein", cd28_protein)

    # Output results for grading verification
    print(f"Task 1 Protein: {storage.read('OriginalProtein')}")
    print(f"Task 5 CD28 Protein: {storage.read('CD28_Protein')}")

    # Task 7: Demonstrate the use of SequenceFactory.
    factory = SequenceFactory()
    test_dna = factory.create_sequence("DNA", "ATGCGT")
    print(f"Factory successfully created: {type(test_dna).__name__}")

    # Demonstration for Task 6: Working with Protein Sequences
    print("\n--- Demonstration for Task 6: Working with Protein Sequences ---")
    protein_seq_str = "MLRLGF"
    protein_obj = factory.create_sequence("PROTEIN", protein_seq_str)
    print(f"Created protein sequence object: {type(protein_obj).__name__} with sequence: {protein_obj.seq}")
    storage.save("TestProtein", protein_obj.seq)
    print(f"Stored and read protein sequence: {storage.read('TestProtein')}")
```
